```python
'''conbined diata frames'''

import re
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import chart_studio.plotly as py
import cufflinks as cf
cf.go_offline()

long_hollidays = pd.read_csv("csv/long_hollidays_clean.csv", sep = ';')
short_hollidays = pd.read_csv("csv/short_hollidays_clean.csv", sep = ';')
long_hollidays = long_hollidays[long_hollidays.Vakantiekenmerken.str.match(pat = '[^%]+')]
# takes out al the seasonal parameters, because these are not in the short holiday dataframe 

short = short_hollidays[["Vliegtuig",'Perioden','Vakantiekenmerken']]#.groupby(['Vakantiekenmerken', 'Perioden']).sum()
long = long_hollidays[["Vliegtuig",'Perioden','Vakantiekenmerken']]#.groupby(['Vakantiekenmerken', 'Perioden']).sum()
#long = long.reset_index()
'''= short.reset_index()
'''
long_hollidays.groupby("Vakantiekenmerken").sum() # L = 212788 S = 53512
short_hollidays.groupby("Vakantiekenmerken").sum() # S = 53512
```

```python
#long_short_vacations = pd.merge(short, long, on=['Vakantiekenmerken','Perioden'], how='left', suffixes=['5', '9'])[30:40]
long_short_vacations = pd.concat([short, long], axis=1, sort=False)
long_short_vacations.columns= ['Vliegtuig5', 'Perioden', 'Vakantiekenmerken', 'Vliegtuig9', 'Perioden9',
       'Vakantiekenmerken9']
long_short_vacations[['Vliegtuig5', 'Perioden', 'Vakantiekenmerken', 'Vliegtuig9',]][40:50]
```

```python
#H0: μ == μ0 
#H1: μ =! μ0 
from scipy.stats import ttest_rel # related sample trest
ttest_rel(list(long_short_vacations['Vliegtuig5']), list(long_short_vacations['Vliegtuig9']))
# Because P is almost 0, we can state H0 is not true, so H1 counts 
```

```python
from scipy.stats import ttest_ind # independed sample test
ttest_ind(long_short_vacations['Vliegtuig5'], long_short_vacations['Vliegtuig9'], equal_var=False)
```
